```python
df_data.columns=["age","blood_pressure","specific_gravity","albumin","sugar",
                 'red_blood_cells','pus_cell','pus_cell_clumps','bacteria','blood_glucose_random','blood_urea','serum_creatinine','sodium',
                 'potassium','haemoglobin','packed_cell_volume','white_blood_cell_count','red_blood_cell_count',
                 'hypertension','diabetes_mellitus','coronary_artery_disease','appetite','peda_edema',
                 'anemia','class']
```

```python
df_data.head()
```

```python
text_columns = ['packed_cell_volume','white_blood_cell_count','red_blood_cell_count']
for i in text_columns:
    print(f'{i} : {df_data[i].dtype}')
```

```python
def convert_text_to_numeric(df_data,column):
    df_data[column] = pd.to_numeric(df_data[column],errors='coerce')
for col in text_columns:
    convert_text_to_numeric(df_data,col)
    print(f'{col} : {df_data[col].dtype}')
```

```python
missing = df_data.isnull().sum()
missing[missing>0].sort_values(ascending=False).head(20)
```

```python
def mean_value_imputation(df_data,column):
    mean_value = df_data[column].mean()
    df_data[column]=df_data[column].fillna(mean_value)

def mode_value_imputation(df_data,column):
    mode_value = df_data[column].mode()[0]
    df_data[column]=df_data[column].fillna(mode_value)

def median_value_imputation(df_data,column):
    median_value = df_data[column].median()
    df_data[column].fillna(median_value)
```

```python
num_cols = [col for col in df_data.columns if df_data[col].dtype!="object"]
obj_cols = [col for col in df_data.columns if df_data[col].dtype=="object"]
```

```python
for col in num_cols:
    mean_value_imputation(df_data,col)
for col in obj_cols:
    mode_value_imputation(df_data,col)
```

```python
missing = df_data.isnull().sum()
missing[missing>0].sort_values(ascending=False).head(20)
```

```python
print(f'diabetes_mellitus :{df_data["diabetes_mellitus"].unique()}')
print(f'coronary_artery_disease :{df_data["coronary_artery_disease"].unique()}')
print(f'class :{df_data["class"].unique()}')
```

```python
df_data["diabetes_mellitus"] = df_data["diabetes_mellitus"].replace(to_replace={
    " yes":"yes","\tno":"no","\tyes":"yes"
})
df_data["coronary_artery_disease"]=df_data["coronary_artery_disease"].replace(to_replace={"\tno":"no"})
df_data["class"] = df_data["class"].replace(to_replace={"ckd\t":"ckd","notckd":"not ckd"})
```

```python
print(f'diabetes_mellitus :{df_data["diabetes_mellitus"].unique()}')
print(f'coronary_artery_disease :{df_data["coronary_artery_disease"].unique()}')
print(f'class :{df_data["class"].unique()}')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
for i in obj_cols:
    df_data[i] = encoder.fit_transform(df_data[i])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_data.drop("class",axis=1)
y=df_data["class"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=25)
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dct = DecisionTreeClassifier()
dct.fit(x_train,y_train)
```

```python
y_pred_dct = dct.predict(x_test)
y_pred_dct
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
models = []
models.append(("Naive Bayes", GaussianNB()))
models.append(("KNN", KNeighborsClassifier()))
models.append(("Random Forest Classifier", RandomForestClassifier()))
models.append(("Decision Tree Classifier", DecisionTreeClassifier()))
models.append(("SVM", SVC()))
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
```

```python
for name,model in models:
    print(name , model)
    print()
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    print(confusion_matrix(y_test,y_pred))
    print("\n")
    print("accuracy: ",accuracy_score(y_test,y_pred))
    print("\n")
    print("precision: ",precision_score(y_test,y_pred))
    print("\n")
    print("recall: ",recall_score(y_test,y_pred))
    print("\n")
    print("f1_score: ",f1_score(y_test,y_pred))
    print("\n")
```
